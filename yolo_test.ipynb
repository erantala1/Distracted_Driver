```python
import torch
import numpy as np
import cv2
from matplotlib import pyplot as plt
import os
import uuid
import time
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
model
```

```python
img = 'https://www.ucsc.edu/wp-content/uploads/2023/09/UCSC_Chancellor_Library.jpg'
test = model(img)
test.show()
```

```python
vid = cv2.VideoCapture(1) #access mac camera
while vid.isOpened():
  val, frame = vid.read() #return value and frame/image read from camera
  detect = model(frame)
  cv2.imshow('test',np.squeeze(detect.render()))
  if cv2.waitKey(10) & 0xFF == ord('q'): #exit by hitting q
    break
vid.release()
cv2.destroyAllWindows()
```

```python
#fine tuning by capturing images from live feed and saving under the two labels

labels = ['awake','sleepy']
num_images = 20
file_path = os.path.join('image_data','images')
label_path = os.path.join('image_data','labels')
vid = cv2.VideoCapture(1)

#create 20 images for each label
for label in labels:
    for i in range(num_images):
        print(f"{label}:{i}")
        val, frame = vid.read()
        
        #creating unique file name for each frame and then saving
        img_name = os.path.join(file_path, label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(img_name,frame)
        cv2.imshow('Image Collection', frame)
        time.sleep(4) #4 second delay
        
        if cv2.waitKey(10) & 0xFF == ord('q'): #exit by hitting q
            break
vid.release()
cv2.destroyAllWindows()
```
